```python
import cv2 # image processing
import numpy as np # provides the container numpy array to hold the 3d image with (h,w,c)
import matplotlib.pyplot as plt # visulize the image


def edge_ded(image_path):

  im = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  # reading your image as a grayscale image

  # vertical edge filter
  ve = np.array([[-1, 0, 1],
                [-2, 0, 2],
                [-1, 0, 1]])

  feature_map = cv2.filter2D(im, cv2.CV_64F, ve)
  # by defult images don't consist of negative number, using cv2.cv_64f we have allowed 64 precision negative numbers

  feature_map_abs = np.absolute(feature_map)

  # normlaize the image back to the range of 0 to 255 to render it as image in python
  feature_map_norm = np.uint8(feature_map_abs / feature_map_abs.max() * 255)

  return feature_map_norm
```
